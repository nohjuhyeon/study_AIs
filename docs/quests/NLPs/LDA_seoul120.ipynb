```python
import pandas as pd
df_seoul120 = pd.read_csv("https://bit.ly/seoul-120-text-csv")
df_seoul120
```

```python
import pandas as pd
import pyLDAvis
import pyLDAvis.lda_model
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

## 자연어 전처리

```python
stopwords = []
```

```python
# 토크나이징 함수 정의
from konlpy.tag import Okt
okt = Okt()
def Tokenizer(raw, pos=["Noun","Verb","Modifier","Adjective"], stopword=stopwords):
    return [
        word for word, tag in okt.pos(
            raw, 
            norm=True,   # normalize 그랰ㅋㅏ -> 그래ㅋㅋ
            stem=True    # stemming 바뀌나->바뀌다
            )
            if len(word) > 1 and tag in pos and word not in stopword

    ]
```

```python
# tokenizing
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfVectorizer = TfidfVectorizer(tokenizer = Tokenizer,max_df=0.95,min_df=2)
```

```python
features = tfidfVectorizer.fit_transform(df_seoul120["내용"])
features.toarray()[:2]
```

```python
import numpy as np 
np.where(features.toarray()[:1] > 0 )
```

```python
latentDirichletAllocation_5=LatentDirichletAllocation(n_components=5,n_jobs=-1)
```

```python
latentDirichletAllocation_5.fit(features)
```

```python
vis_5 = pyLDAvis.lda_model.prepare(latentDirichletAllocation_5,features,tfidfVectorizer)
vis_5
```

```python
pyLDAvis.enable_notebook()
components_5_display = pyLDAvis.display(vis_5)
components_5_display
```
