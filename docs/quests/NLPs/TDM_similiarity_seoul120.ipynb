## 자연어 전처리

```python
stopwords = []
```

```python
# 토크나이징 함수 정의
from konlpy.tag import Okt
okt = Okt()
def Tokenizer(raw, pos=["Noun","Verb"], stopword=stopwords):
    return [
        word for word, tag in okt.pos(
            raw, 
            norm=True,   # normalize 그랰ㅋㅏ -> 그래ㅋㅋ
            stem=True    # stemming 바뀌나->바뀌다
            )
            if len(word) > 1 and tag in pos and word not in stopword

    ]
```

```python
# tokenizing
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfVectorizer = TfidfVectorizer(tokenizer = Tokenizer)
```

```python
features = tfidfVectorizer.fit_transform(df_seoul120["내용"])
features.toarray()[:2]
```

## 유사도 확인

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
result_list = cosine_similarity(features.toarray()[1772:1773],features.toarray())
series_result = pd.Series(data =result_list[0])
series_result
```

```python
df_seoul120['유사도'] = series_result
df_seoul120.sort_values(by='유사도',ascending = False)[:5]
```
