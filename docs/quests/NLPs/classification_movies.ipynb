```python
from mecab import MeCab
mecab = MeCab()
```

```python
df_review['document'] = df_review['document'].str.upper()
```

```python
mecab.pos(df_review['document'][149768])
```

```python
for j in range(len(df_review)):
    document_sentence = ""
    for i in mecab.pos(df_review['document'][j]):
        if i[1] == "VA" or i[1] == "NNG" or i[1] == "NNG" or i[1] == "VV" or i[1] == "MAG" or i[1] == "SY" or i[1] == "IC" or i[1] == "SL":
            document_sentence = document_sentence + (i[0]) + " "
    df_review.loc[j,'document_clean']=document_sentence
df_review
```

```python
condition_non = 'document_clean ==""'
df_review.query(condition_non)
```

```python
condition_a = 'document.str.contains("재미|재밌|재밓|재밋")'
conditions = f'{condition_non} and {condition_a}'
df_review.query('document_clean=="별로"')
```

```python
index_num = df_review.query(conditions).index
df_review.loc[index_num,'document_clean'] += '재미 '
```

```python
condition_non_a = f"{condition_non} or document_clean == '재미 '"
df_review.query(condition_non_a)
```

```python
condition_b =  'document.str.contains("별로|별루")'
```

```python
conditions = f"{condition_non_a} and {condition_b}"
df_review.query(conditions)
```

```python
index_num = df_review.query(conditions).index
df_review.loc[index_num,'document_clean'] += '별로 '
```

```python
condition_non_b = f"{condition_non_a} or document_clean == '별루 '"
df_review.query(condition_non_b)
```

```python
df_review['document_clean'] = df_review['document_clean'].str.replace('별루','별로',regex=True)
```

```python
df_review.query('document_clean ==""')
```

## 토큰화

```python
# tokenizing
from sklearn.feature_extraction.text import TfidfVectorizer
# tfidfVectorizer = TfidfVectorizer(tokenizer=None)
tfidfVectorizer= TfidfVectorizer(tokenizer=None,ngram_range=(1,2))
```

```python
tfidfVectorizer.fit(df_review['document_clean'])
```

```python
df_train_set = df_review[:100000]
df_train_set
```

```python
df_test_set = df_review[100000:]
df_test_set
```

```python
test_target = df_test_set['label']
test_features = tfidfVectorizer.transform(df_test_set['document_clean'])
```

```python
train_target = df_train_set['label']
train_features = tfidfVectorizer.transform(df_train_set['document_clean'])
```

```python
from sklearn.ensemble import RandomForestClassifier
randomForestClassifier=RandomForestClassifier(n_jobs=-1)
```

```python
randomForestClassifier.fit(train_features,train_target)
```
