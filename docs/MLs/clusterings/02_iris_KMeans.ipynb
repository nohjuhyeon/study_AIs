- 군집화

```python
from sklearn import datasets
iris_dataset =datasets.load_iris()
```

```python
type(iris_dataset),iris_dataset.keys() # bunch : dictionary
```

```python
feature_train = iris_dataset['data']
feature_train[:2]
```

## 모델

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
```

```python
model.fit(feature_train)
```

```python
model.predict([[5.1,3.5,1.4,0.2]])
```

```python
iris_dataset['target'][:1]
```

## 평가

```python
model.cluster_centers_ # model.predict([[5.1,3.5,1.4,0.2]])의 값은 model.cluster_centers_의 두번째와 유사하다
```

```python
n_cluster_list = [2,3,4,5,6,7,8,9,10]

result_list =[]
for n_cluster in n_cluster_list:
  n_model = KMeans(n_clusters=n_cluster)
  n_model.fit(feature_train)
  print(n_model.inertia_)
  result_list.append(n_model.inertia_)
  pass
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x=n_cluster_list,y=result_list)
sns.lineplot(x=n_cluster_list,y=result_list)
plt.show() # 점을 기준으로 값이 급격히 바뀌는 구간을 n_cluster의 개수로 선택하는게 좋음(이 선택이 정확하지는 않음)
```
