## 군집화(Clustering)

### 데이터

```python
import numpy as np
```

```python
X = np.array([[1, 2], [1, 4], [10, 0],[1, 0],[10, 2], [10, 4] ])
# X = np.array([[1, 2], [1, 4], [1, 0],[10, 2], [10, 4], [10, 0]])
```

```python
type(X),X.shape
```

### 모델

```python
from sklearn import cluster
```

```python
kMeans = cluster.KMeans(n_clusters=2)  ## K : 중앙개수의 값
kMeans
```

```python
kMeans.fit(X)
```

### 평가

```python
kMeans.labels_    # y 값
```
