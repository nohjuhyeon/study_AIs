## 데이터


```python
from sklearn import datasets
iris_dataset =datasets.load_iris()
```

```python
type(iris_dataset),iris_dataset.keys() # bunch : dictionary
```

```python
feature_train = iris_dataset['data']
feature_train[:2]
```

```python
feature_train.shape
```

```python
# creat imbalance dataset
imbalance_features = iris_dataset['data'][:70]
imbalance_target = iris_dataset['target'][:70]
imbalance_features.shape, imbalance_target.shape
```

```python
from collections import Counter
Counter(imbalance_target)
```

## resampling

### Oversampling (많은 개수 맞춤)


```python
from imblearn.over_sampling import SMOTE
overSampling = SMOTE(sampling_strategy=0.8)
feature_over_sample, target_over_sample= overSampling.fit_resample(imbalance_features, imbalance_target)
feature_over_sample.shape, target_over_sample.shape
```

```python
Counter(target_over_sample)
```

### UnderSampling(적은 개수 맞춤)

```python
# Oversampling (많은 개수 맞춤)
from imblearn.under_sampling import NearMiss
underSampling = NearMiss(sampling_strategy=0.8)
feature_under_sample, target_under_sample= underSampling.fit_resample(imbalance_features, imbalance_target)
feature_under_sample.shape, target_under_sample.shape
```

```python
Counter(target_under_sample)
```

### Over와 Under 결합 Sampling

```python
from imblearn.combine import SMOTEENN
combineSampling = SMOTEENN(sampling_strategy=0.5)
feature_combine_sample, target_combine_sample=combineSampling.fit_resample(imbalance_features, imbalance_target)
feature_combine_sample.shape, target_combine_sample.shape
```

```python
Counter(target_combine_sample)
```
