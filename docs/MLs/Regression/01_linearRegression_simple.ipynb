## 회귀 예측(Regression)

- 데이터
- 모델
- 평가

### 데이터 관련

```python
from sklearn import datasets
diabetes_X,diabetes_y = datasets.load_diabetes(return_X_y= True)
```

```python
type(diabetes_X),type(diabetes_y),diabetes_X.shape,diabetes_y.shape
```

### 모델

```python
from sklearn import linear_model    # 선형모델
```

```python
linearRegression = linear_model.LinearRegression() # 학습 장소 선택
```

```python
linearRegression.fit(diabetes_X,diabetes_y) # 학습 진행
```

### 평가

```python
diabetes_X[:2],diabetes_y[:2]
```

```python
linearRegression.predict(diabetes_X[:2])
```

### 학습 공식 확인

```python
linearRegression.coef_,linearRegression.coef_.shape,linearRegression.intercept_
```

```python
diabetes_X[:1]
```

```python
(linearRegression.coef_ * diabetes_X[:1]).sum() + linearRegression.intercept_
```
