```python
import pandas as pd
import pyLDAvis
import pyLDAvis.lda_model
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.datasets import fetch_20newsgroups
# 데이터 로드
data = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'))
documents = data.data
```

```python
documents[:1]
```

### TfidfVector
- max_df : 0.95( 전체 단어 중 95% 이상 들어간 단어 빼기)
- min_df : 2(전체 단어 중 2개 이하로 나온 단어 빼기

```python
tfidfVectorizer=TfidfVectorizer(stop_words='english',max_df=0.95,min_df=2)
```

```python
tfidf = tfidfVectorizer.fit_transform(documents)
tfidf.toarray()[:1].shape
```

```python
import numpy as np 
np.where(tfidf.toarray()[:1] > 0 )
```

```python
latentDirichletAllocation=LatentDirichletAllocation(n_components=5,n_jobs=-1)
```

```python
latentDirichletAllocation.fit(tfidf)
```

```python
vis = pyLDAvis.lda_model.prepare(latentDirichletAllocation,tfidf,tfidfVectorizer)
vis
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.display(vis)
```
